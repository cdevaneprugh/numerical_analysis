# Basic Numerical Derivative Functions

```python
# derivative functions

def forward_diff(x0, h):
    fprime = ( f(x0 + h) - f(x0) ) / h
    return fprime

def backward_diff(x0, h):
    fprime = ( f(x0) - f(x0 - h) ) / h
    return fprime

def center_diff(x0, h):
    fprime = ( f(x0 + h) - f(x0 - h) ) / (2 * h)
    return fprime
```

# Example Over Even Grid

```python
# define function
def f(x):
    return x**2
```

```python
# list of x values to plot/differentiate over
x = np.arange(-10,11, dtype=float)

# corresponding y vlaues for plot
y = []
for i in x:
    y.append(f(i))
    
# differentiate for each item in list
h = 0.01
fprime = []
for i in x:
    fprime.append(center_diff(i, h))
```

```python
plt.plot(x, y, 'o--', color = 'k', label='f (x)')
plt.plot(x, fprime, 'o--', color = 'r', label="f '(x)")
plt.title('Function and First Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```

# Uneven Grid

```python
def uneven_grid(x, h1, h2):    
        
    # define terms
    term_1 = ( h1 * f(x+h))  / ( h2 * (h1+h2) )
    term_2 = ( (h1-h2) * f(x))  / (h2*h1)
    term_3 = ( h2 * f(x-h))  / ( h1 * (h1+h2) )
        
    # calculate & add to list
    fprime = term_1 - term_2 - term_3 
        
    return fprime
```

```python
def f(x):
    return x**3

# list of x values to plot/differentiate over
x = np.arange(-10,11, dtype=float)

# corresponding y vlaues for plot
y = []
for i in x:
    y.append(f(i))

# differentiate
h = 0.01
dydx = uneven_grid(x, h, h)

plt.plot(x, y, 'o--', color = 'k', label='f (x)')
plt.plot(x, dydx, 'o--', color = 'r', label="f '(x)")
plt.title('Function and First Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.show()
```
